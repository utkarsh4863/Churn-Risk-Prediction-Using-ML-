### Test dataset

```python
tf = pd.read_csv("datasets/test.csv")
tf.head()
```

```python
# preprocessing
# filling categorical missing column
tf['region_category']= tf['region_category'].fillna(tf['region_category'].mode()[0])
tf['preferred_offer_types']= tf['preferred_offer_types'].fillna(tf['preferred_offer_types'].mode()[0])

# handle numerical missing column
tf['points_in_wallet']= tf['points_in_wallet'].fillna(tf['points_in_wallet'].mean())

# drop one column
tf.drop('complaint_status', axis=1, inplace=True)

# Exploratory data analysis
tf["joining_day"] = pd.to_datetime(tf.joining_date, format="%Y-%m-%d").dt.day
tf["joining_month"] = pd.to_datetime(tf.joining_date, format="%Y-%m-%d").dt.month
tf["joining_year"] = pd.to_datetime(tf.joining_date, format="%Y-%m-%d").dt.year

tf.drop(["joining_date"], axis=1, inplace=True)

# Extract hour, minutes and second
tf[['last_visit_time_hour','last_visit_time_minutes','last_visit_time_seconds']] = tf['last_visit_time'].astype(str).str.split(':', expand=True).astype(int)

# Now we can drop Arrival_Time as it is of no use
tf.drop(["last_visit_time"], axis = 1, inplace = True)

# Handling categorical data
gender = pd.get_dummies(tf[['gender']], drop_first=True, dtype='uint8')
region = pd.get_dummies(tf[['region_category']], drop_first=True, dtype='uint8')
membership = pd.get_dummies(tf[['membership_category']], drop_first=True, dtype='uint8')
referal = pd.get_dummies(tf[['joined_through_referral']], drop_first=True, dtype='uint8')
offer = pd.get_dummies(tf[['preferred_offer_types']], drop_first=True, dtype='uint8')
medium = pd.get_dummies(tf[['medium_of_operation']], drop_first=True, dtype='uint8')
internet = pd.get_dummies(tf[['internet_option']], drop_first=True, dtype='uint8')
discount = pd.get_dummies(tf[['used_special_discount']], drop_first=True, dtype='uint8')
application = pd.get_dummies(tf[['offer_application_preference']], drop_first=True, dtype='uint8')
complaint = pd.get_dummies(tf[['past_complaint']], drop_first=True, dtype='uint8')
feedbacks = pd.get_dummies(tf[['feedback']], drop_first=True, dtype='uint8')

data_test = pd.concat([tf, gender, region, membership, referal, offer, medium, internet, discount, application, complaint, feedbacks ], axis = 1)

data_test.drop(["customer_id", "Name", "gender", "security_no", "region_category", "membership_category", "joined_through_referral", "referral_id", "preferred_offer_types", "medium_of_operation", "internet_option", "used_special_discount", "offer_application_preference", "past_complaint", "feedback", "avg_frequency_login_days" ], axis = 1, inplace = True)

data_test.head()
```

```python
# Fix negative class label
y_train = y_train.replace(-1, 0)
y_test = y_test.replace(-1, 0)
```

```python
# XGBoost clasifier
from xgboost import XGBClassifier
xg = XGBClassifier(objective='binary:logistic')
xg.fit(x_train, y_train)
```

```python
# checking training accuracy
print("training accuracy is : ", xg.score(x_train, y_train)*100)
```

```python
# checking accuracy of test dataset
print("testing accuracy is : ", xg.score(x_test, y_test)*100)
```

## Hyperparameter Tuning


* Choose following method for hyperparameter tuning
    1. **RandomizedSearchCV** --> Fast
    2. **GridSearchCV**
* Assign hyperparameters in form of dictionery
* Fit the model
* Check best paramters and best score

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid={
    'learning_rate':[1,0.5,0.1,0.01],
    'max_depth':[3,5,10,20],
    'n_estimators':[10,50,100,200]
}
```

```python
grid = RandomizedSearchCV(XGBClassifier(objective='binary:logistic'), param_grid, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
# After tuning


xg = XGBClassifier(objective='binary:logistic', n_estimators =100, max_depth=5, learning_rate =0.1)
xg.fit(x_train, y_train)
```

```python
# Predict
y_pred = xg.predict(x_test)





from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Evaluation
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

# Overall accuracy
acc = accuracy_score(y_test, y_pred)
print(f"Overall Accuracy: {acc:.4f}")
```

```python
# checking accuracy of test dataset
print("testing accuracy is : ", xg.score(x_test, y_test)*100)
```
